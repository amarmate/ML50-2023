```python
import random 
import numpy 
import pandas 
import matplotlib.pyplot as plt
```

```python
skill = numpy.random.normal(0,10,500)
practice = numpy.random.normal(0,10,500)
success = (skill + practice)/2 + numpy.random.normal(-2.5,2.5,500)

# Scale success to be between 0 and 10 (MinMaxScaler)
success = (success - min(success))/(max(success) - min(success)) * 10

X = pandas.DataFrame({'skill':skill, 'practice':practice, 'success':success})
```

```python
# Calculate the correlation between skill and success
print("Correlation between skill and practice (not conditioning on success): ", X.corr().iloc[0,1])
print("Correlation between skill and practice (conditioning on success): ", X[X.success > 9].corr().iloc[0,1])

X.plot.scatter(x='skill', y='practice', c='success', colormap='viridis')
```
